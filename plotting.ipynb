# Plotting

## With 60% stragglers

### Mu = 0

```python
fig = plt.figure()
ax = fig.add_subplot(111)
linestyles = ["-", "-", "-", "-", ":", ":", ":", ":"]
colors = [
    "tab:red",
    "tab:orange",
    "tab:green",
    "tab:blue",
    "tab:red",
    "tab:orange",
    "tab:green",
    "tab:blue",
]
labels = [
    "FedAvg",
    "FedAvg + Adagrad",
    "FedAvg + Adam",
    "FedAvg + Yogi",
    "FedMed",
    "FedMed + Adagrad",
    "FedMed + Adam",
    "FedMed + Yogi",
]
for idx, hist in enumerate(
    [
        hist_sgd_06,
        hist_adagrad_06,
        hist_adam_06,
        hist_yogi_06,
        hist_sgd_06_fedmed,
        hist_adagrad_06_fedmed,
        hist_adam_06_fedmed,
        hist_yogi_06_fedmed,
    ]
):
    rounds, values = zip(*hist["accuracy"])
    plt.plot(
        np.asarray(rounds)[::10],
        np.asarray(values)[::10],
        linestyle=linestyles[idx],
        color=colors[idx],
        label=labels[idx],
    )

plt.ylim([0.80, 1.001])
plt.title("MNIST CNN Non-IID with 60% stragglers and mu=0").set_fontsize(14)
plt.xlabel("Communication Rounds").set_fontsize(14)
plt.ylabel("Test Accuracy").set_fontsize(14)
plt.legend(bbox_to_anchor=(1.04, 1), borderaxespad=0)
for label in ax.yaxis.get_ticklabels()[::2]:
    label.set_visible(False)
for label in ax.xaxis.get_ticklabels():
    label.set_fontsize(14)
for label in ax.yaxis.get_ticklabels():
    label.set_fontsize(14)
plt.yticks(rotation=90, va="center")
ax.margins(x=0.03)
ratio = 1.05
ax.tick_params(axis="both", which="major", width=1, length=10)
xleft, xright = ax.get_xlim()
ybottom, ytop = ax.get_ylim()
ax.set_aspect(abs((xright - xleft) / (ybottom - ytop)) * ratio)
fig.savefig(
    "centralized_metrics_06_fedmed.png",
    bbox_inches="tight",
    pad_inches=0.4,
)
plt.close()
```
